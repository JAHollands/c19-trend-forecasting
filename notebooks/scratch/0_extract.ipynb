```python
import pandas as pd
from pathlib import Path
```

```python
url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/worldwide-aggregate.csv"
df = pd.read_csv(url)

df.head()
df.info()
df.describe(include="all")
df["Date"].min(), df["Date"].max()
```

check cols as expected

```python
expected_cols = {"Date", "Confirmed", "Recovered", "Deaths", "Increase rate"}
assert expected_cols.issubset(df.columns), f"Missing columns: {expected_cols - set(df.columns)}"
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
```

handle null values identified

```python
df["Increase rate"] = df["Increase rate"].fillna(0)
df["Increase rate"].head(3)
```
